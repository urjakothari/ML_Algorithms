```python
#separating test set and train set

import hashlib
def test_set_check(identifier,test_ratio,hash):
    return bytearray(hash(np.int64(identifier)).digest())[-1]  < 51
def split_train_test_by_id(data,test_ratio,id_column,hash):
    ids=data[id_column]
    in_test_set=ids.apply(lambda id_:test_set_check(id_,test_ratio,hash))
    return in_test_set
```

# Logistic Regression

```python
#Logistic Regression

def sigmoid(z):
    return 1.0 / (1 + np.exp(-z))

#Regularized gradient function
def Gradient(theta, X, y):
    m, n = X.shape
    theta = theta.reshape((n, 1))
    y = y.reshape((m, 1))
    h = sigmoid(X.dot(theta))
    return ( X.T.dot(y-h))

#Optimal theta 
eta=0.002
def logisticRegression(X, y, theta,iterations):
    for i in range(iterations):
        #result = op.minimize(fun = regCostFunction, x0 = theta, args = (X, y),method = 'TNC', jac = regGradient)
        gradient=Gradient(theta,X,y)
        #print(gradient)
        #print("\n")
        theta=theta+eta*gradient
    return theta
```

```python
import time
#Training
iterations=5000
n=7
theta =np.matrix( np.zeros((n + 1,1)))
#One vs all
y_train=np.array(y_train)
start_time=time.clock()
optTheta = logisticRegression(X_train, y_train, theta,iterations)
end_time=time.clock()
print(optTheta)
```

```python
#Predictions
P = sigmoid(X_test.dot(optTheta)) #probability for each flower
P=P.as_matrix()
P = np.where(P > 0.5, 1, 0)
from sklearn.metrics import accuracy_score
print ("Test Accuracy using Logistic Regression(Gradient Descent)", accuracy_score(y_test, P) * 100 , '%')
print ("Training time : ",(end_time-start_time))
```

```python
from sklearn.metrics import precision_recall_fscore_support
#function returns precision, recall,fscore
Precision, Recall, Fscore, Support = precision_recall_fscore_support(y_test,P, average='binary')
print("Precision",Precision)
```
